```python
#%% md
# ex4_scientific2
## Nội dung bài
# Hiển thị 2 đồ thị sin(x), cos(x) trong khoảng x ∈ [-π, π]

#%% md
### Import thư viện
#%%
import numpy as np
import matplotlib.pyplot as plt

#%% md
### Sinh dữ liệu hàm số
#%%
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)  # sinh 256 điểm trong [-π, π]
C, S = np.cos(X), np.sin(X)

print(f'+ X: |X|={len(X)}, x in [{np.min(X):.2f}, {np.max(X):.2f}], 10 samples: {X[:10]}')
print(f'+ Cos(X): |C|={len(C)}, C in [{np.min(C):.2f}, {np.max(C):.2f}], 10 samples: {C[:10]}')
print(f'+ Sin(X): |S|={len(S)}, S in [{np.min(S):.2f}, {np.max(S):.2f}], 10 samples: {S[:10]}')

#%% md
### Hiển thị đồ thị
#%%
plt.figure(figsize=(8,6))

# vẽ cos(x) và sin(x)
plt.plot(X, C, color="blue", linewidth=2.5, linestyle="-", label="cos(x)")
plt.plot(X, S, color="red", linewidth=2.5, linestyle="-", label="sin(x)")

# giới hạn trục x
plt.xlim(X.min() * 1.1, X.max() * 1.1)
plt.xticks(
    [-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
    [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$']
)

# giới hạn trục y
plt.ylim(-1.1, 1.1)
plt.yticks([-1, 0, +1], [r'$-1$', r'$0$', r'$+1$'])

# thêm tiêu đề + chú thích
plt.title("Đồ thị sin(x) và cos(x)")
plt.legend(loc="upper left")

# hiển thị
plt.show()

#%% md
# Kết thúc
```
