```python
# import
import pandas as pd
import pandas_datareader.data as web
import matplotlib.pyplot as plt
import mplfinance as mpf 
import seaborn as sns
import datetime as dt
%matplotlib inline
from IPython.display import display
import yfinance as yf
```

```python
# Tải dữ liệu cổ phiếu AAPL
# pandas_datareader hiện không hoạt động với Yahoo Finance

start = dt.datetime(2016, 1, 1)
end = dt.datetime.now()
df = yf.Ticker('AAPL').history(start=start, end=end, auto_adjust=False) 

print(f'{start} -> {end}')
```

```python
display(df.head())
display(df.tail())
```

```python
df.loc['2025-08-29']
```

```python
df.to_csv('AAPL.csv')
df1 = pd.read_csv('AAPL.csv', index_col='Date')
```

```python
display(df1)
```

```python
df['Adj Close'].plot(legend=True, figsize=(12,8), title='AAPL', label = 'Adj Close')
plt.show()
```

```python
plt.figure(figsize=(12, 9))

top = plt.subplot2grid((12, 9), (0, 0), rowspan=10, colspan=9)
bottom = plt.subplot2grid((12, 9), (10, 0), rowspan=2, colspan=9)

top.plot(df.index, df['Adj Close'], color='blue')
bottom.bar(df.index, df['Volume'])

top.axes.get_xaxis().set_visible(False)
top.set_title('AAPL')
top.set_ylabel('Adj Close')
bottom.set_ylabel('Volume')
plt.show()
```
