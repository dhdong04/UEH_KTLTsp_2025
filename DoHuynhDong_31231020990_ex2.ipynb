```python
#%% md
# ex2_scientific1

## Ná»™i dung bÃ i
# Hiá»ƒn thá»‹ ngáº«u nhiÃªn N=50 hÃ¬nh trÃ²n:
# + TÃ¢m c_i = (x_i, y_i) ngáº«u nhiÃªn vá»›i x_i, y_i âˆˆ [0, 1]
# + MÃ u sáº¯c color_i ngáº«u nhiÃªn vá»›i color_i âˆˆ [0, 1]
# + Diá»‡n tÃ­ch area_i = Ï€ * r_i^2 vá»›i r_i âˆˆ [0, 1]

#%% md
### Import thÆ° viá»‡n

#%%
import numpy as np
import matplotlib.pyplot as plt

#%% md
### Sinh ngáº«u nhiÃªn dá»¯ liá»‡u hÃ¬nh trÃ²n

#%%
# Äáº·t seed cho tÃ¡i láº­p
np.random.seed(42)

N = 50
x = np.random.rand(N)          
y = np.random.rand(N)         
colors = np.random.rand(N)     
r = np.random.rand(N)         
area = np.pi * (r ** 2)      

#%% md
### Kiá»ƒm tra nhanh dá»¯ liá»‡u

#%%
print(f'x: len={len(x)}, range=[{x.min():.2f}, {x.max():.2f}], samples={x[:10]}')
print(f'y: len={len(y)}, range=[{y.min():.2f}, {y.max():.2f}], samples={y[:10]}')
print(f'colors: len={len(colors)}, range=[{colors.min():.2f}, {colors.max():.2f}], samples={colors[:10]}')
print(f'r: len={len(r)}, range=[{r.min():.2f}, {r.max():.2f}], samples={r[:10]}')
print(f'area=Ï€r^2: len={len(area)}, range=[{area.min():.4f}, {area.max():.4f}], samples={area[:10]}')

#%% md
### Hiá»ƒn thá»‹ Ä‘á»“ thá»‹
plt.figure(figsize=(6, 6))
plt.scatter(x, y, s=area * 4000, c=colors, alpha=0.6)
plt.title('Ngáº«u nhiÃªn 50 hÃ¬nh trÃ²n (tÃ¢m, mÃ u, r âˆˆ [0,1]; area=Ï€rÂ²)')
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.xlabel('x')
plt.ylabel('y')
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
