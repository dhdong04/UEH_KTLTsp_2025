```python
#%% md
# ex2_scientific1

## Nội dung bài
# Hiển thị ngẫu nhiên N=50 hình tròn:
# + Tâm c_i = (x_i, y_i) ngẫu nhiên với x_i, y_i ∈ [0, 1]
# + Màu sắc color_i ngẫu nhiên với color_i ∈ [0, 1]
# + Diện tích area_i = π * r_i^2 với r_i ∈ [0, 1]

#%% md
### Import thư viện

#%%
import numpy as np
import matplotlib.pyplot as plt

#%% md
### Sinh ngẫu nhiên dữ liệu hình tròn

#%%
# Đặt seed cho tái lập
np.random.seed(42)

N = 50
x = np.random.rand(N)          
y = np.random.rand(N)         
colors = np.random.rand(N)     
r = np.random.rand(N)         
area = np.pi * (r ** 2)      

#%% md
### Kiểm tra nhanh dữ liệu

#%%
print(f'x: len={len(x)}, range=[{x.min():.2f}, {x.max():.2f}], samples={x[:10]}')
print(f'y: len={len(y)}, range=[{y.min():.2f}, {y.max():.2f}], samples={y[:10]}')
print(f'colors: len={len(colors)}, range=[{colors.min():.2f}, {colors.max():.2f}], samples={colors[:10]}')
print(f'r: len={len(r)}, range=[{r.min():.2f}, {r.max():.2f}], samples={r[:10]}')
print(f'area=πr^2: len={len(area)}, range=[{area.min():.4f}, {area.max():.4f}], samples={area[:10]}')

#%% md
### Hiển thị đồ thị
plt.figure(figsize=(6, 6))
plt.scatter(x, y, s=area * 4000, c=colors, alpha=0.6)
plt.title('Ngẫu nhiên 50 hình tròn (tâm, màu, r ∈ [0,1]; area=πr²)')
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.xlabel('x')
plt.ylabel('y')
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
